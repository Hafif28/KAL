## SPL 3 Variabel (Eliminasi Gaussian)

### Contoh SPL 3 Variabel

Sistem persamaan yang akan diselesaikan:

$$
\begin{cases}
x + y + z = 6 \\
2x + 3y + z = 10 \\
x + 2y + 3z = 13
\end{cases}
$$

### Matriks Augmented

$$
\begin{bmatrix}
1 & 1 & 1 & | & 6 \\
2 & 3 & 1 & | & 10 \\
1 & 2 & 3 & | & 13
\end{bmatrix}
$$

### Langkah Eliminasi Gaussian

1. Eliminasi elemen di bawah pivot pertama.
2. Eliminasi elemen di bawah pivot kedua.
3. Ubah pivot terakhir menjadi 1.
4. Lakukan substitusi balik untuk memperoleh solusi.

```python
import numpy as np

# Matriks augmented SPL 3 variabel
A = np.array([
    [1, 1, 1, 6],
    [2, 3, 1, 10],
    [1, 2, 3, 13]
], dtype=float)

print('Matriks Awal:')
print(A)
print('-' * 40)

n = len(A)

# Eliminasi Gaussian
for i in range(n):
    pivot = A[i][i]

    # Normalisasi pivot
    if pivot != 0:
        A[i] = A[i] / pivot

    print(f'Langkah Normalisasi Baris {i+1}')
    print(A)

    # Eliminasi ke bawah
    for j in range(i + 1, n):
        faktor = A[j][i]
        A[j] = A[j] - faktor * A[i]

        print(f'Eliminasi Baris {j+1}')
        print(A)

    print('-' * 40)

print('Bentuk REF:')
print(A)

# Back Substitution
x = np.zeros(n)

for i in range(n - 1, -1, -1):
    x[i] = A[i][-1] - np.sum(A[i][i+1:n] * x[i+1:n])

print('\nSolusi SPL:')
print(f'x = {x[0]:.2f}')
print(f'y = {x[1]:.2f}')
print(f'z = {x[2]:.2f}')
```

## SPL 4 Variabel (Eliminasi Gaussian)

### Contoh SPL 4 Variabel

$$
\begin{cases}
2x_1 + 3x_2 - x_3 + 4x_4 = 1 \\
x_1 - 2x_2 + 2x_3 - 3x_4 = 2 \\
3x_1 + x_2 + x_3 + 2x_4 = 3 \\
-x_1 + 4x_2 + x_3 - 2x_4 = 4
\end{cases}
$$

### Matriks Augmented

$$
\begin{bmatrix}
2 & 3 & -1 & 4 & | & 1 \\
1 & -2 & 2 & -3 & | & 2 \\
3 & 1 & 1 & 2 & | & 3 \\
-1 & 4 & 1 & -2 & | & 4
\end{bmatrix}
$$

Metode yang digunakan:

1. Eliminasi Gaussian untuk membentuk REF.
2. Eliminasi Gauss-Jordan untuk membentuk RREF.
3. Mengambil solusi dari kolom terakhir.

```python
import numpy as np

# Matriks augmented SPL 4 variabel
A = np.array([
    [2, 3, -1, 4, 1],
    [1, -2, 2, -3, 2],
    [3, 1, 1, 2, 3],
    [-1, 4, 1, -2, 4]
], dtype=float)

print('Matriks Awal:')
print(A)
print('-' * 50)

n = len(A)

# Eliminasi Gaussian
for i in range(n):

    # Tukar baris jika pivot = 0
    if A[i][i] == 0:
        for k in range(i + 1, n):
            if A[k][i] != 0:
                A[[i, k]] = A[[k, i]]
                print(f'Tukar Baris {i+1} dengan Baris {k+1}')
                print(A)
                break

    # Normalisasi pivot
    pivot = A[i][i]
    A[i] = A[i] / pivot

    print(f'Normalisasi Baris {i+1}')
    print(A)

    # Eliminasi ke bawah
    for j in range(i + 1, n):
        faktor = A[j][i]
        A[j] = A[j] - faktor * A[i]

        print(f'Eliminasi Baris {j+1}')
        print(A)

    print('-' * 50)

print('Bentuk REF:')
print(A)

# Eliminasi ke atas (Gauss-Jordan)
for i in range(n - 1, -1, -1):
    for j in range(i - 1, -1, -1):
        faktor = A[j][i]
        A[j] = A[j] - faktor * A[i]

print('\nBentuk RREF:')
print(A)

# Solusi akhir
solusi = A[:, -1]

print('\nSolusi SPL:')
for i, nilai in enumerate(solusi):
    print(f'x{i+1} = {nilai:.4f}')
```

# Soal Determinan Matriks

Hitunglah determinan matriks menggunakan metode ekspansi baris:

$$
\det(A) = \sum_{k=1}^{n} (-1)^{i+k} a_{ik} M_{ik}
$$

dengan:

$$
M_{ik} = \det(A_{ik})
$$

dan $A_{ik}$ adalah submatriks yang diperoleh dengan menghapus baris ke-$i$ dan kolom ke-$k$.
