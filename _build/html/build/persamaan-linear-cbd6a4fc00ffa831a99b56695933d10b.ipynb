## Sistem Persamaan Linear (System of Linear Equations)

Sistem Persamaan Linear (SPL) adalah kumpulan dari dua atau lebih persamaan linear yang melibatkan variabel-variabel yang sama. Persamaan dikatakan **linear** jika setiap variabelnya berpangkat satu dan tidak ada perkalian antar variabel.

**Bentuk Umum:**
Sebuah persamaan linear dengan $n$ variabel $x_1, x_2, \dots, x_n$ dapat dituliskan dalam bentuk:
$$a_1x_1 + a_2x_2 + \dots + a_nx_n = b$$

Dimana:
- $a_1, a_2, \dots, a_n$ adalah koefisien (bilangan riil).
- $x_1, x_2, \dots, x_n$ adalah variabel atau anu (unknowns).
- $b$ adalah konstanta di ruas kanan.

**Tujuan Utama:**
Mencari nilai-nilai dari setiap variabel (seperti $x, y, z$) yang jika disubstitusikan ke dalam **seluruh** persamaan dalam sistem tersebut, akan menghasilkan pernyataan yang benar secara bersamaan.

## Representasi Sistem Persamaan Linear

Untuk memudahkan perhitungan, terutama pada sistem yang besar, kita dapat menyajikannya dalam beberapa bentuk:

### a. Bentuk Skalar (Persamaan Biasa)
Menuliskan setiap persamaan secara eksplisit baris demi baris.
Contoh:
$$\begin{cases} 2x + 3y = 5 \\ 4x - y = 3 \end{cases}$$

### b. Bentuk Persamaan Matriks
Menyederhanakan sistem menjadi operasi perkalian matriks:
$$A\mathbf{x} = \mathbf{b}$$

Dimana:
- $A$ adalah **Matriks Koefisien**: Berisi angka-angka di depan variabel.
- $\mathbf{x}$ adalah **Vektor Variabel**: Berisi variabel yang ingin dicari nilainya.
- $\mathbf{b}$ adalah **Vektor Konstanta**: Berisi nilai di ruas kanan persamaan.

### c. Augmented Matrix (Matriks Diperluas)
Menggabungkan matriks $A$ dan vektor $\mathbf{b}$ ke dalam satu matriks tunggal $[A|b]$ untuk mempermudah proses eliminasi.
Contoh:
$$\left[ \begin{array}{cc|c} 2 & 3 & 5 \\ 4 & -1 & 3 \end{array} \right]$$

## Himpunan Penyelesaian (Solusi)

Solusi dari sistem persamaan linear adalah sekumpulan nilai variabel yang memenuhi semua persamaan di dalam sistem tersebut. Setiap sistem persamaan linear hanya memiliki tiga kemungkinan jumlah solusi:

1.  **Solusi Tunggal (Unique Solution):** Hanya ada satu set nilai yang memenuhi.
2.  **Solusi Tak Hingga (Infinite Solutions):** Terdapat banyak sekali (tak terbatas) kombinasi nilai yang memenuhi.
3.  **Tidak Ada Solusi (No Solution):** Tidak ada satu pun nilai yang bisa memenuhi semua persamaan sekaligus.

Sistem yang memiliki setidaknya satu solusi disebut **Konsisten**, sedangkan sistem yang tidak memiliki solusi disebut **Tidak Konsisten (Inconsistent)**.

## Interpretasi Geometris

Visualisasi membantu kita memahami mengapa sebuah sistem bisa memiliki solusi atau tidak melalui pendekatan geometris:

- **Sistem 2 Variabel ($x, y$):** Setiap persamaan merepresentasikan sebuah **garis lurus** pada bidang Cartesian 2D.
- **Sistem 3 Variabel ($x, y, z$):** Setiap persamaan merepresentasikan sebuah **bidang datar** dalam ruang Euclidean 3D.

### Analisis Visual dan Interaktif
Kita dapat menggunakan alat seperti **GeoGebra** untuk melihat bagaimana garis atau bidang ini berinteraksi:

1.  **Satu Titik Potong (Unique Solution):** Garis atau bidang bertemu di tepat satu titik koordinat.
2.  **Garis/Bidang Sejajar (No Solution):** Garis atau bidang tidak pernah bertemu karena memiliki kemiringan yang sama namun posisi berbeda.
3.  **Garis/Bidang Berhimpit (Infinite Solutions):** Garis atau bidang saling menumpuk sepenuhnya karena persamaannya sebenarnya identik.

## Metode Penyelesaian Utama

Ada beberapa pendekatan teknis untuk menemukan nilai variabel dalam SPL:

### 1. Eliminasi Gauss (Gaussian Elimination)
Metode sistematis yang mengubah matriks diperluas menjadi **Eselon Baris** menggunakan Operasi Baris Elementer (OBE):
- Menukar posisi dua baris.
- Mengalikan baris dengan konstanta tak nol.
- Menjumlahkan kelipatan satu baris ke baris lainnya.
*Ini adalah metode yang paling efisien untuk komputasi komputer.*

### 2. Metode Substitusi
Mengisolasi satu variabel di satu persamaan, lalu memasukkannya ke persamaan lainnya. Sangat efisien untuk sistem kecil (2 atau 3 variabel).

### 3. Aturan Cramer (Cramer's Rule)
Menggunakan perbandingan **Determinan** matriks.
- Rumus: $x_i = \frac{\det(A_i)}{\det(A)}$
- Keterbatasan: Hanya bisa digunakan jika matriks $A$ adalah persegi dan determinannya tidak nol. Tidak efisien untuk sistem skala besar.

## Sistem Persamaan Linear Homogen

Sistem disebut homogen jika semua konstanta di ruas kanan adalah nol:
$$A\mathbf{x} = \mathbf{0}$$

**Karakteristik Utama:**
1.  **Selalu Konsisten:** Pasti memiliki minimal satu solusi.
2.  **Solusi Trivial:** Solusi di mana semua variabel bernilai nol ($x_1=0, x_2=0, \dots$).
3.  **Solusi Non-Trivial:** Jika sistem memiliki solusi selain nol, maka otomatis sistem tersebut memiliki **tak hingga solusi**. Ini terjadi jika matriks $A$ bersifat singular (determinannya nol).
